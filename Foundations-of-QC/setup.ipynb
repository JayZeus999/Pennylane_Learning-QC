```python
import pennylane as qp

qml = qp

from pennylane import numpy as np
```

```python
# Specify a device on which the quantum circuit will run. 
dev = qp.device("default.qubit", wires=2)

# The pairing of a quantum circuit with a device is called a QNode.
# @qp.qnode(dev)

def my_first_quantum_function(theta):
    qp.RX(theta, wires=0)
    qp.PauliY(wires=1)
    qp.Hadamard(wires=0)
    qp.Hadamard(wires=1)

    return qp.state()
```

```python
# Distinguish the function from the QNode
my_first_qnode = qp.QNode(my_first_quantum_function, dev)
```

```python
print(my_first_qnode(np.pi/4))
```

The list above represents the amplitudes of the state in the computational basis 

in this order. 

{|00>, |01>, |10>, |11>}


```python
dev_c_t = qp.device("default.qubit", wires = ["target", "control"])

@qp.qnode(dev_c_t)

def create_entangled():
    qp.Hadamard(wires = "control")
    qp.CNOT(wires = ["control", "target"])

    return qp.state()
```

```python
print(create_entangled)
```

Quantum functions that do not return anything can be used as subcircuits—smaller pieces that build up to a larger circuit.

```python
def subcircuit_1(angle):
    
    qp.RX(angle, wires = 0)
    qp.PauliY(wires = 1)
```

```python
def subcircuit_2():
    
    qp.Hadamard(wires = 0)
    qp.CNOT(wires = [0, 1])
```

```python
def full_circuit(theta, phi):

    subcircuit_1(theta)
    subcircuit_2()
    subcircuit_1(phi)
```

```python
theta = 0.3
phi = 0.2
```

```python
print(qp.draw(full_circuit)(theta, phi))
```
